```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, chi2

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve

import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
# Log transforming Amounts to remove skewness

df['Amount']=np.log1p(df['Amount'])
```

```python
# Visualizing Distribution for the Amounts

plt.plot(figsize=(10,6))

sns.histplot(df['Amount'],kde=True, bins=30)

plt.title('Distribution of Log transformed Amounts')

plt.xlabel('Log Transformed Amounts')

plt.ylabel('Frequency')

plt.grid(True)
```

```python
# Select feature to scale

Features_to_scale = ['V1','V2','V3','V4']

# Scaling Features

scaler = StandardScaler()

df_scaled = df.sample(n=2000, random_state = 42).copy()

df_scaled[Features_to_scale] = scaler.fit_transform(df_scaled[Features_to_scale])

sns.pairplot(df_scaled[['V1','V2','V3','V4','Class']], hue = 'Class', diag_kind='kde')
```

```python
# Seperate Feature and Target

x = df.drop(columns = ['id','Class'], axis = 1)

y = df['Class']

# Train test split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 7)
```

```python
# SelectKBest for feature selection

x_train_copy = x_train.copy()

x_test_copy = x_test.copy()

# MixMaxScaler to make all features non-negative

scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform(x_train_copy)

x_test_scaled = scaler.transform(x_test_copy)

# Scaled data for feature selection

select_k_best = SelectKBest(chi2, k = 10)

x_train_selected = select_k_best.fit_transform(x_train_scaled, y_train)

x_test_selected = select_k_best.transform(x_test_scaled)

selected_features = x.columns[select_k_best.get_support()]

selected_features
```

```python
# Train Random Forest Model

model = RandomForestClassifier(random_state = 7)

model.fit(x_train_selected, y_train)
```

```python
# Evaluating Model

y_pred = model.predict(x_test_selected)

# Acccuracy Score

accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
```

```python
# Model Parameters

print(classification_report(y_test, y_pred))
```

```python
# Save Model

joblib.dump(model, "random_forest_model.pkl")
print("Model saved!")
```
